```python
data['Ram'] = data['Ram'].str.replace('GB',"").astype('int32')
```

```python
data['Weight'] = data['Weight'].str.replace('kg','').astype('float32')
```

```python
data.head()
```

```python
data.corr()['Price_euros']
```

```python
data['Company'].value_counts(0)
```

```python
def add_to_other(inp):
    if inp == 'Samsung' or inp == 'Razer' or inp == 'Mediacom' or inp == 'Microsoft' or inp == 'Xiaomi' or inp == 'Vero' or inp == 'Chuwi' or inp == 'Google' or inp == 'Fujitsu' or inp == 'LG' or inp == 'Huawei' :
        return 'Other'
    else:
        return inp
```

```python
data['Company'] = data['Company'].apply(add_to_other)
data['Company'].value_counts()
```

```python
data.head()
```

```python
data['ScreenResolution'].unique()
```

```python
data['Touchscreen'] = data['ScreenResolution'].apply(lambda x:1 if 'Touchscreen' in x else 0)
```

```python
data['IPS'] = data['ScreenResolution'].apply(lambda x:1 if 'IPS' in x else 0)
```

```python
data.head()
```

```python
data['Cpu'] = data['Cpu'].apply(lambda x:' '.join(x.split()[0:3])) 
```

```python
data["Cpu"].value_counts()
```

```python
def set_processor(name):
    if name == 'Intel Core i7' or name == 'Intel Core i5' or name == 'Intel Core i3':
        return name
    else:
        if name.split()[0] == 'AMD':
            return 'AMD'
        else:
            return 'Other'
```

```python
data['Cpu'] = data['Cpu'].apply(set_processor)
```

```python
data['Cpu'].value_counts()
```

```python
data.head()
```

```python
data['Gpu'] = data['Gpu'].apply(lambda x:' '.join(x.split()[0:1]))
data.head()
```

```python
data = data[data['Gpu']!='ARM']
data["Gpu"].value_counts()
```

```python
data["OpSys"].value_counts()
```

```python
def add_os(inp):
    if inp == 'Windows 10' or inp == 'Windows 7' or  inp == 'Windows 10 S':
        return 'Windows'
    elif inp == 'Linux' or inp == 'No OS':
        return inp
    elif inp == 'macOS' or inp == 'Mac OS X':
        return 'Mac'
    else:
        return 'Other'
```

```python
data["OpSys"] = data["OpSys"].apply(add_os)
data["OpSys"].value_counts()
```

```python
data = data.drop(columns=['laptop_ID','Inches','Product','ScreenResolution'])
```

```python
data.head()
```

```python
data = pd.get_dummies(data)
data.head()
```

```python
data.info()
```

# Model Building

```python
X = data.drop(columns=['Price_euros'])
y = data['Price_euros']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
X_train.shape, X_test.shape
```

# Model Training

```python
def model_acc(model):
    model.fit(X_train, y_train)
    acc = model.score(X_test, y_test)
    print(str(model)+ ' --> ' +str(acc))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
model_acc(lr)

from sklearn.linear_model import Lasso
lasso = Lasso()
model_acc(lasso)

from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
model_acc(dt)

from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
model_acc(rf)
```

# Hyperparameter Tunning

```python
from sklearn.model_selection import GridSearchCV
parameters = {'n_estimators':[20, 40, 60, 80, 100], 'criterion':['squared_error','absolute_error','poisson']}
grid_obj = GridSearchCV(estimator=rf, param_grid=parameters)
grid_fit = grid_obj.fit(X_train, y_train)
best_model = grid_fit.best_estimator_
best_model.score(X_test, y_test)
```

# Save Model

```python
import pickle
with open('predictor.pickle', 'wb') as file:
    pickle.dump(best_model,file)
```
